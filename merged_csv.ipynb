```python
# Dependencies and Setup
import pandas as pd

# csv files
gdp = "Output/cleaned_gdp.csv"
cpi = "Output/cleaned_cpi.csv"

# Read csv files
gdp_data = pd.read_csv(gdp)
cpi_data = pd.read_csv(cpi)

gdp_data
```

```python
cpi_data
```

```python
merged_gdp_cpi = pd.merge(gdp_data, cpi_data, how="right", on=["Month"])
```

```python
merged_gdp_cpi.head()
```

```python
merged_gdp_cpi = merged_gdp_cpi.rename(columns={"Precent_GDP":"Percent GDP",
                                                "Precent_CPI":"Percent CPI"})
```

```python
merged_gdp_cpi
```

```python
# csv files
commodity = "Output/cleaned_commodity.csv"
unemployment = "Output/cleaned_unemployment rate.csv"

# Read csv files
commodity_data = pd.read_csv(commodity)
unemployment_data = pd.read_csv(unemployment)

commodity_data
```

```python
unemployment_data
```

```python
merged_commodity_unemployment = pd.merge(commodity_data, unemployment_data, how="left", on=["Month"])
```

```python
merged_commodity_unemployment
```

```python
merged_commodity_unemployment = merged_commodity_unemployment.rename(columns={
                                                "Seasonally adjusted (%)":"Seasonally adjusted unemployment (%)"})
```

```python
merged_commodity_unemployment
```

```python
merged_4_data = pd.merge(merged_gdp_cpi, merged_commodity_unemployment, how="left", on=["Month"])
```

```python
merged_4_data.head()
```

```python
# csv files
interest_cash_exchange = "Output/cleaned_interest_cash_exchange.csv"
household_finance = "Output/cleaned_household_finance.csv"
```

```python
# Read csv files
interest_cash_exchange_data = pd.read_csv(interest_cash_exchange)
household_finance_data = pd.read_csv(household_finance)
```

```python
interest_cash_exchange_data.head()
```

```python
household_finance_data.head()
```

```python
household_finance_data = household_finance_data.rename(columns={"Date":"Month"})
```

```python
household_finance_data.head()
```

```python
merged_5_data = pd.merge(merged_4_data, household_finance_data, how="left", on=["Month"])
```

```python
merged_5_data
```

```python
interest_cash_exchange_data.head()
```

```python
merged_data = pd.merge(merged_5_data, interest_cash_exchange_data, how="left", on=["Month"])
```

```python
merged_data.to_csv("Merged/merged_data.csv", index=False, header=True)
```
