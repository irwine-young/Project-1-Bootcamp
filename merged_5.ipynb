```python
# Dependencies and Setup
import pandas as pd

# csv files
gdp = "Output/cleaned_gdp.csv"
cpi = "Output/cleaned_cpi.csv"

# Read csv files
gdp_data = pd.read_csv(gdp)
cpi_data = pd.read_csv(cpi)

gdp_data
```

```python
cpi_data
```

```python
merged_gdp_cpi = pd.merge(gdp_data, cpi_data, how="right", on=["Month"])
```

```python
merged_gdp_cpi.head()
```

```python
merged_gdp_cpi = merged_gdp_cpi.rename(columns={"Precent_GDP":"Percent GDP",
                                                "Precent_CPI":"Percent CPI"})
```

```python
merged_gdp_cpi
```

```python
# csv files
commodity = "Output/cleaned_commodity.csv"
unemployment = "Output/cleaned_unemployment rate.csv"

# Read csv files
commodity_data = pd.read_csv(commodity)
unemployment_data = pd.read_csv(unemployment)

commodity_data
```

```python
unemployment_data
```

```python
merged_commodity_unemployment = pd.merge(commodity_data, unemployment_data, how="left", on=["Month"])
```

```python
merged_commodity_unemployment
```

```python
merged_commodity_unemployment = merged_commodity_unemployment.rename(columns={
                                                "Seasonally adjusted (%)":"Seasonally adjusted unemployment (%)"})
```
